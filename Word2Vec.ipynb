```python
!pip install gensim
import gensim
import pandas as pd
```

```python
df = pd.read_json("Sports_and_Outdoors_5.json", lines=True, encoding='utf-8')
```

```python
df = df.reviewText.apply(gensim.utils.simple_preprocess)
```

```python
m = gensim.models.Word2Vec(
    window=4,
    workers = 4,
    min_count=2,
    epochs= 4
)
```

```python
m.build_vocab(df)
```

```python
m.train(df, total_examples=m.corpus_count, epochs=m.epochs)
```

```python
m.save("./k")
```

```python
k = gensim.models.Word2Vec.load("./k")
```
